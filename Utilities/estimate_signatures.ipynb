# Signature sizes assuming no seed tree is employed

```python
import matplotlib.pyplot as plt

lambda_val = 128

s_vals = []
sign_vals = []

for s in range(2,11):
    t = ceil(lambda_val/log(s,2))
    sign_size = t*(lambda_val/s + 2*lambda_val*(1-1/s))/8000
    s_vals.append(s)
    sign_vals.append(sign_size)

plt.plot(s_vals, sign_vals, color = "blue")
plt.grid()
plt.ylabel("Signature size (kB)")
plt.xlabel("s (Number of Public keys)")
plt.show()
```
